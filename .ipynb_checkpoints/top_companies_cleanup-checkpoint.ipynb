```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import load_model
```

```python
aapldf = pd.read_csv('AAPL_prediction.csv', index_col=0, names=['Apple'])
disdf = pd.read_csv('DIS_prediction.csv', index_col=0, names=['Disney'])
gedf = pd.read_csv('GE_prediction.csv', index_col=0, names=['Gen_Elec'])
googdf = pd.read_csv('GOOG_prediction.csv', index_col=0, names=['Google'])
ibmdf = pd.read_csv('IBM_prediction.csv', index_col=0, names=['IBM'])
jpmdf = pd.read_csv('JPM_prediction.csv', index_col=0, names=['JP_Morgan'])
strbxdf = pd.read_csv('SBUX_prediction.csv', index_col=0, names=['Starbucks'])
tgtdf = pd.read_csv('TGT_prediction.csv', index_col=0, names=['Target'])
ualdf = pd.read_csv('UAL_prediction.csv', index_col=0, names=['Untd_Air'])
aapldf.head()
```

```python
aapldf['Date'] = 'Date'
aapldf['Date']= 'Date'
disdf['Date']= 'Date'
gedf['Date']= 'Date'
googdf['Date']= 'Date'
ibmdf['Date']= 'Date'
jpmdf['Date']= 'Date'
strbxdf['Date']= 'Date'
tgtdf['Date']= 'Date'
ualdf['Date']= 'Date'

aapldf
```

```python
aapldf1= aapldf.reset_index()
disdf1= disdf.reset_index()
gedf1= gedf.reset_index()
googdf1= googdf.reset_index()
ibmdf1= ibmdf.reset_index()
jpmdf1= jpmdf.reset_index()
strbxdf1= strbxdf.reset_index()
tgtdf1= tgtdf.reset_index()
ualdf1= ualdf.reset_index()
strbxdf1
```

```python
aapldf1['Date'] = pd.date_range(start='5/6/2020', periods=len(aapldf), freq='D')
disdf1['Date'] = pd.date_range(start='5/6/2020', periods=len(disdf), freq='D')
gedf1['Date'] = pd.date_range(start='5/6/2020', periods=len(gedf), freq='D')
googdf1['Date'] = pd.date_range(start='5/6/2020', periods=len(googdf), freq='D')
ibmdf1['Date'] = pd.date_range(start='5/6/2020', periods=len(ibmdf), freq='D')
jpmdf1['Date'] = pd.date_range(start='5/6/2020', periods=len(jpmdf), freq='D')
strbxdf1['Date'] = pd.date_range(start='5/6/2020', periods=len(strbxdf), freq='D')
tgtdf1['Date'] = pd.date_range(start='5/6/2020', periods=len(tgtdf), freq='D')
ualdf1['Date'] = pd.date_range(start='5/6/2020', periods=len(ualdf), freq='D')
ualdf1.head()
```

```python
ualdf1.set_index('Date').join(tgtdf1.set_index('Date'))
```

```python
dfs = [aapldf1,disdf1,gedf1,googdf1,ibmdf1,jpmdf1,strbxdf1,tgtdf1,ualdf1]
dfs = [df.set_index('Date') for df in dfs] 
dfs1 = dfs[0].join(dfs[1:])
dfs1.round(2)
```

```python
dfs1.to_csv('prediction_table.csv')
```
